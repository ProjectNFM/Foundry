```python
from temporaldata import Data
import h5py

path = "/home/mila/s/sobralm/Foundry/data/processed/peterson_brunton_pose_trajectory_2022/AJILE12_P01_20000102_ses3_pose_trajectories.h5"

with h5py.File(path, "r") as f:
    data = Data.from_hdf5(f, lazy=True)
    data.active_vs_inactive_trials.materialize()
    data.splits.materialize()
```

```python
data.active_vs_inactive_trials
active = data.active_vs_inactive_trials.behavior_labels == "Active"
inactive = data.active_vs_inactive_trials.behavior_labels == "Inactive"

active_trials = data.active_vs_inactive_trials.select_by_mask(active)
inactive_trials = data.active_vs_inactive_trials.select_by_mask(inactive)

print(sum(active_trials.end - active_trials.start))
print(sum(inactive_trials.end - inactive_trials.start))
```

```python
data.splits.active_vs_inactive_fold_0_train
active_train = data.splits.active_vs_inactive_fold_0_train.behavior_labels == "Active"
inactive_train = data.splits.active_vs_inactive_fold_0_train.behavior_labels == "Inactive"

active_trials_fold_0_train = data.splits.active_vs_inactive_fold_0_train.select_by_mask(active_train)
inactive_trials_fold_0_train = data.splits.active_vs_inactive_fold_0_train.select_by_mask(inactive_train)

print(sum(active_trials_fold_0_train.end - active_trials_fold_0_train.start))
print(sum(inactive_trials_fold_0_train.end - inactive_trials_fold_0_train.start))
```
